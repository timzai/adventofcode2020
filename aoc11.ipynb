<p>Now, you just need to model the people who will be arriving shortly. Fortunately, people are entirely predictable and always follow a simple set of rules. All decisions are based on the <em>number of occupied seats</em> adjacent to a given seat (one of the eight positions immediately up, down, left, right, or diagonal from the seat). The following rules are applied to every seat simultaneously:</p>
<ul>
<li>If a seat is <em>empty</em> (<code>L</code>) and there are <em>no</em> occupied seats adjacent to it, the seat becomes <em>occupied</em>.</li>
<li>If a seat is <em>occupied</em> (<code>#</code>) and <em>four or more</em> seats adjacent to it are also occupied, the seat becomes <em>empty</em>.</li>
<li>Otherwise, the seat's state does not change.</li>
</ul>

```python
import numpy as np
from itertools import product
import copy

def day11():
    # Load in the data file. 0 = floor, 1 = empty seat, 2 = taken seat, -1 = out of bounds
    d = np.array([[int(y) for y in list(x)] for x in open("data/day11.txt").read().replace(".","0").replace("L","1").split()])
    
    # add out of bounds
    d = np.pad(d,((1,1),(1,1)),'constant', constant_values=((-1,-1),(-1,-1)))
    
    # get coordinates of all seats
    seats = list(zip(*np.nonzero(d == 1)))
    
    # generate 8 directions
    all_directions = [(x,y) for x,y in product(range(-1,2),repeat=2) if (x,y)!=(0,0)]

    # this is for finding the coordinates of visible neighbors (for part 2)
    def look_further(i,j,di,dj):
        if d[i,j] == 0:
            return look_further(i+di,j+dj,di,dj)
        elif d[i,j] == 1:
            return (i,j)
        else: # out of bound
            return None

    # we do a one off search for the 2 neighbor types for all seats.  we only keep track coordinates with seats
    neighbors = {} # part 1 neighbors
    visibles = {} # part 2 neighbors
    for i,j in seats:
        neighbors[(i,j)] = [(i+k,j+l) for k,l in all_directions if d[i+k,j+l]==1] # adjacent ones
        visibles[(i,j)] = [look_further(i+k,j+l,k,l) for k,l in all_directions if look_further(i+k,j+l,k,l) != None]

    def solve_part(d,neighbor,t):
        o = np.empty((0,0))
        while not np.array_equal(o,d):
            o=copy.deepcopy(d)
            for i,j in seats:
                occupied_neighbors = sum([1 for k,l in neighbor[(i,j)] if o[k,l] == 2])
                if occupied_neighbors == 0:
                    d[i,j] = 2
                else:
                    if occupied_neighbors >= t:
                        d[i,j] = 1
        return np.count_nonzero(d==2)

    return (solve_part(copy.deepcopy(d),neighbors,4), solve_part(copy.deepcopy(d),visibles,5))

day11()
```
